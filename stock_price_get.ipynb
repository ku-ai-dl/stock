## 全銘柄の株価を取得

```python
import yfinance as yf
import time

from pandas_datareader import data as pdr
#pandas_datareaderオーバーライド
yf.pdr_override()
```

```python
start = pd.to_datetime('2000-01-01') 
start
```

```python
datetime.date.today()
```

```python
def get_brand(code):
    try:
        start = pd.to_datetime('2000-01-01') #取得したい日付のはじめ
        end = pd.to_datetime(datetime.date.today())　#取得したい日付の最後(今日にしている)
        stock = (str(code) + ".T")
        data = pdr.get_data_yahoo(stock, start=start, end=end)
        
        df = data.reset_index()
        df = df.assign(code=code)
        df = df.drop(['Adj Close'], axis=1)
        
        print(code)

    except (ValueError, IndexError, AttributeError, KeyError):
        return None
    
    return df

def brands_generator(code_range):
    cols = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'code']
    df = pd.DataFrame(index=[], columns=cols)
    
    for code in code_range:
        
        #生成した証券コードをスクレイピング関数へ渡す          
        brand = get_brand(code)
    
        #情報が取得できていれば、情報を結合していく
        if brand is not None:
            df = pd.concat([df, brand]).reset_index(drop=True)
        
        #1秒間プログラムを停止する(スクレイピング負荷対策)
        time.sleep(0.1)
    
    return df
```

```python
# df = brands_generator(range(1301,1310))
df = brands_generator(range(1301,9999))
```

## DBを作成

### ①priceテーブル

```python
import sqlite3
import pandas as pd
```

```python
dbname = ('jp_stock.db')#データベース名.db拡張子で設定
con = sqlite3.connect(dbname, isolation_level=None)#データベースを作成、自動コミット
cur = con.cursor() #データベースに接続する

cur.execute("create table if not exists price(date integer, Open real, High real, Low real, Close real, Volume integer, code integer)")
```

```python
#テーブル確認
pd.read_sql_query('SELECT name FROM sqlite_master WHERE TYPE="table"', con)
```

```python
df.shape
```

```python
df.to_sql('price', con, if_exists='replace', index=None, method='multi', chunksize=5000)
```

```python
"""
https://www.jpx.co.jp/markets/statistics-equities/misc/01.htmlをそのまま使用する
"""

import requests
dls = "https://www.jpx.co.jp/markets/statistics-equities/misc/tvdivq0000001vg2-att/data_j.xls"
resp = requests.get(dls)

output = open('data_j.xls', 'wb')
output.write(resp.content)
output.close()
```

```python
pd_br = pd.read_excel('data_j.xls')
```

```python
pd_br.head(3)
```

```python
#前処理
pd_br = pd_br.drop(['日付'], axis=1)
```

```python
pd_br.head(1)
```

```python
#名前を変更
pd_br = pd_br.rename(columns={'コード': 'code', '33業種コード':'業種コード_33', '33業種区分':'業種区分_33',
                             '17業種コード':'業種コード_17', '17業種区分':'業種区分_17'})
```

```python
pd_br.head(1)
```

```python
#ハイフンをNanに置き換える ※でないとエラーになるので
pd_br = pd_br.replace('-', np.nan)
```

```python
#数字が最初にくるカラムはエラーになってしまう・・
cur.execute("""create table if not exists Brands(code integer PRIMARY KEY, 銘柄名 text NOT NULL, 市場・商品区分 text,
            業種コード_33 integer, 業種区分_33 text, 業種コード_17 integer, 業種区分_17 text, 規模コード integer, 規模区分 text)""")
```

```python
pd_br.to_sql('Brands', con,  if_exists='replace', index=None, method='multi')
```
